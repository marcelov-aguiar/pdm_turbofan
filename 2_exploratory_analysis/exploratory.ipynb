```python
def add_rul(g):
    g['RUL'] = [max(g['time'])] * len(g)
    #print([max(g['time'])])
    g['RUL'] = g['RUL'] - g['time']
    del g['unit_number']
    return g.reset_index()
```

```python
df_data_train.groupby('unit_number').apply(add_rul).reset_index()
```
